## mpg데이터 분석

```python
mpg=pd.read_csv('mpg.csv')
mpg
```

```python
mpg_a=mpg.query('displ <= 4')
mpg_a.head()
```

```python
mpg_b = mpg.query('displ >=5')
mpg_b.head()
```

```python
mpg_a['hwy'].mean()
```

```python
mpg_b['hwy'].mean()
```

```python
if mpg_a['hwy'].mean()>mpg_b['hwy'].mean():
    print('배기량 4이하 자동차의 고속도로 연비가 높다')
elif mpg_a['hwy'].mean()<mpg_b['hwy'].mean():
    print('배기량 4이상 자동차의 고속도로 연비가 높다')
```

```python
exam = pd.read_csv('..\Day1\exam.csv')
```

```python
exam.sort_values('math')
```

```python
# ascending default 올림차순
# descending 내림차순 

#수학점수로 내림차순
exam.sort_values('math', ascending=False)
```

```python
exam.sort_values(['nclass', 'math'], ascending=[True, False])
```

```python
#map fillter
```

```python
exam.assign(total = lambda x : x['math']+x['english']+x['science']).head(3)
```

```python
exam.head(3)
```

```python
exam.assign(total = lambda x : x['math']+x['english']+x['science'],
    mean = lambda x: x['total']/3).sort_values('total', ascending=False).head(5)
```

## Lab3

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg_new=mpg.copy()
```

```python
mpg_new = mpg_new.assign(total=mpg_new['cty']+mpg_new['hwy'])
mpg_new.head()
```

```python
mpg_new=mpg_new.assign(mean=mpg_new['total']/2)
mpg_new.head(3)
```

```python
mpg_new.sort_values('mean', ascending=False).head(3)
```

```python
mpg.assign(total = lambda x : x['cty']+x['hwy'],
          mean = lambda x : x['total']/2).sort_values('mean', ascending=False).head(3)
```
