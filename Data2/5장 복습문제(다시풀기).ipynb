## 5장 복습문제

```python
import pandas as pd
mpg=pd.read_csv('mpg.csv')
import numpy as np
```

```python
mpg= mpg.assign(total = mpg['cty'] + mpg['hwy'])
mpg.head()
```

```python
#요7약 통계량에서 25% 75% 값 확인하기

mpg.describe()
```

```python
mpg['grade']=np.where(mpg['total']>23.5,'A',
                     np.where(mpg['total']>26.5, 'B','C'))
mpg.head()
```

```python
mpg_grade=mpg['grade'].value_counts()
```

```python
mpg_grade.plot.bar(rot=1)
```
