```python
import pandas as pd
exam=pd.read_csv('..\Day1\exam.csv')
exam
```

```python
#aggregate 합계 총액  반별로 묶고(groupby) 합계(agg) 사용
# agg (추가하는 변수명 = ('원래 변수명', '함수명'))

exam.agg(math_mean=('math','mean'),
         eng_mean = ('english','mean'),
         학생수 = ('id', 'count'))
```

### 집단별로 요약 통계량 구하기

```python
# 반 별로 그룹화 하고 그룹별로 각 변수에 대한 평균 값 구하기


exam.groupby('nclass').mean()
```

```python
# 변수이름을 mean_math로 하고 기본적으로 nclass가 인덱스가 됨

exam.groupby('nclass').agg(mean_math = ('math','mean'))
```

```python
# 인덱스를 nclass 기본으로 만들지 않음

exam.groupby('nclass', as_index=False).agg(mean_math = ('math','mean'))
```

```python
exam.groupby('nclass')\
.agg(수학평균 =('math','mean'),
     수학총합 = ('math','sum'))
```

```python
exam.groupby('nclass')\
    .agg(mean_math=('math','mean'),
         sum_math=('math','sum'),
         median_math=('math', 'median'),
         n = ('math','count'))
```

```python
import pandas as pd
fruit=pd.DataFrame({'Num':[123,456,789,1010,1112],
                   'fruit':['apple','banana','cherry','lemon','peach']})

grade = pd.DataFrame({'Num':[123,789,1011],
                     'Grade':['A','B','C']})

fruit
```

```python
#left join  왼쪽 데이터프레임을 기준으로 조인한다. 오른쪽 데이터 프레임에 없는 값은 nan으로 
pd.merge(fruit,grade,on='Num', how = 'left')
```

```python
#오른쪽 기준으로

pd.merge(fruit, grade,on='Num', how='right')
```

```python
# 교집합 양쪽에 공통으로 있는 값만

pd.merge(fruit, grade, on='Num', how='inner')
```

```python
# 모든 값이 나타나도록

pd.merge(fruit, grade, on='Num', how ='outer')
```

## Lab4 분석도전

```python
midwest = pd.read_csv('midwest.csv')
midwest.head(3)
```

```python
# '전체 인구 대비 미성년 인구 백분율'인 파생변수 ratio 추가

midwest['ratio'] = (midwest['poptotal']-midwest['popadults'])/midwest['poptotal']*100

midwest.head(3)
```

```python
midwest[['county','state','ratio']].head(3)
```

* 미성년 비율이 높은 상위 5개 지역의 인구 백분율을 출력

```python
midwest.sort_values('ratio',ascending=False).head(5)[['state','county','ratio']]
```

* 분류 기준표에 따라 미성년 비율 등급 변수를 추가하고, 각 등급에 몇개 지역이 있는지 알아보기

```python
import numpy as np
```

```python
# ratio(전체 인구대비 미성년 인구 백분율)이 40이상-> large
#                                           30 이상-> middle
#                                           나머지 -> small


midwest['grade'] = np.where(midwest['ratio'] >=40 , 'large',
                   np.where(midwest['ratio']>=30, 'middle','small'))

midwest.head(3)[['county','ratio','grade']]
```

```python
midwest.groupby('grade').agg(count_grade = ('grade', 'count'))
```
