```python
# 컬럼명 평균기온-> temp_mean으로 변경
temp_new=temp.copy()
# temp_new
temp_new = temp_new.rename(columns={temp.columns[2]:'temp_mean'})
```

```python
temp_new
```

```python
temp_new=temp_new.rename(columns={temp.columns[3]:'temp_min'})
```

```python
temp_new=temp_new.rename(columns={temp.columns[4]:'temp_max'})
```

```python
temp_new.info()
```

```python
temp_new[['temp_max']].max()
```

```python
temp_new.query('temp_max==39.6')
```

```python
# 날짜 앞에 \t지우기


temp_new['날짜'] = temp_new['날짜'].str.replace('\t','')
```

```python
temp_new.head()
```

```python
temp_new.info()
```

```python
temp_new['날짜']=pd.to_datetime(temp_new['날짜'])
```

```python
temp_new.info()
```

```python
temp_new['year']=temp_new['날짜'].dt.year
```

```python
temp_new.head()
```

```python
temp_new['month'] = temp_new['날짜'].dt.month
```

```python
temp_new.head()
```

```python
temp_new['day']=temp_new['날짜'].dt.day
```

```python
temp_new.head()
```

```python
temp_new['diff']=temp_new['temp_max']-temp_new['temp_min']
```

```python
temp_new.head()
```

```python
temp_new['diff'].isna().sum()
```

```python
temp_new = temp_new.dropna( subset={'diff'})
```

```python
temp_new['diff'].isna().sum()
```

## 일교차가 가장 컷던 날

```python
temp_new.sort_values('diff', ascending=False).head()
```

```python
temp_new['diff'].max()
```

```python
# 데이터프레임에서 결측치 개수 확인

pd.isna(temp_new).sum()
```

## 월별 평균 일교차가 큰 달을 알아보기

```python
temp_new.groupby('month')\
    .agg(diff_mean=('diff','mean'))\
    .sort_values('diff_mean', ascending=False)
```

```python
temp_new.groupby('month')\
        .agg(diff_mean=("diff", "mean"))\
        .sort_values('month')
```

## 월별 평균 기온 알아보기

```python
temp_new.groupby('month')\
    .agg(monthly=('temp_mean','mean'))
```

## 1월 기온만 추출하기

```python
temp_jan=temp_new.query('month==1')
```

```python
temp_jan
```

```python
temp_jan_mean = temp_jan.groupby('year', as_index=False)\
                        .agg(temp_jan_mean=('temp_mean','mean'))
temp_jan_mean
```

```python
temp_jan_mean.plot.line(x='year', y='temp_jan_mean')
```
